### Calculate some variables
*  GDD (base 0), GDD (base 5), Chill days (base 5)
*  GDD accumulations will begin on DOY 1
*  Chill day accumulations will be roughly September 1-December 31 and January 1-March 31
*  All use MEAN daily temperature

```python
# Empty arrays to concatenate cumulative these new data for all years (row-wise)
cumgdd0    = np.empty(0); # Cumulative GDD, Base 0
cumgdd5    = np.empty(0); # Cumulative GDD, Base 5
chill_days = np.empty(0); # Chill days
chill_mon  = np.zeros((np.size(np.unique(prism_yrs)),12))*np.nan

# Dummy variable to store DOY information
prism_doy = np.zeros(prism_yrs.shape)

# Loop through each year and calculate GDDs and chill days
for i_yr in enumerate(np.unique(prism_yrs)):
    #print(i_yr)

    # Find index location for all data for current year
    yr_locs = np.where(prism_yrs==i_yr[1])[0]
    
    # DOY Vector
    prism_doy[yr_locs] = np.arange(1,yr_locs.size+1)
    
    # Pull out all temperature data for this  current year
    curr_tmean = tmean[yr_locs]

    # GDD (base 0) Calculations----------------------------------------------------------------------------------------
    
    # Set baseline for GDD Calc
    gdd_base = 0.0
    
    # Find locations of GDDs in base temperature series
    i_GDD     = np.where(curr_tmean>gdd_base)

    # Calculate  GDDs
    gdd        = np.zeros(yr_locs.shape); 
    gdd[i_GDD] = curr_tmean[i_GDD]-gdd_base

    # Calculate Cumulate Sum and Concatenate (row wise) All Years
    cumgdd0 = np.concatenate((cumgdd0,scipy.cumsum(gdd)),axis=0)

    # GDD (base 5) Calculations----------------------------------------------------------------------------------------
    
    # Set baseline for GDD Calc
    gdd_base = 5.0
    
    # Find locations of GDDs in base temperature series
    i_GDD = np.where(curr_tmean>gdd_base)

    # Calculate  GDDs
    gdd        = np.zeros(yr_locs.shape); 
    gdd[i_GDD] = curr_tmean[i_GDD]-gdd_base

    # Calculate Cumulate Sum and Concatenate (row wise) All Years
    cumgdd5 = np.concatenate((cumgdd5,scipy.cumsum(gdd)),axis=0)
    
    # Chill Day Calculations-------------------------------------------------------------------------------------------

    # Set baseline for ChillCalc
    chill_base = 5.0
    
    # Find locations of GDDs in base temperature series
    i_CHILL = np.where(curr_tmean<=chill_base)
    
    # Calculate Chill Days (binary, 1 or 0)
    chill          = np.zeros(yr_locs.shape)
    chill[i_CHILL] = 1.0

    # Calculate Cumulate Sum and Concatenate (row wise) All Years
    chill_days = np.concatenate((chill_days,chill),axis=0)

    # Monthly Chill Day Totals-----------------------------------------------------------------------------------------
    
    # DOY indices for each month this year
    days_yr = np.zeros((12,1))
    for i_mon in enumerate(mons):
        days_yr[i_mon[0]] = calendar.monthrange(np.int64(i_yr[1]),i_mon[1])[1]
    cum_days_yr = np.cumsum(days_yr)
    
    # Vector for all days of the year
    alldays_yr = np.arange(1,cum_days_yr[-1]+1)
    
    # Loop through and calculate for each month
    for i_mon in enumerate(mons):
    
        # Find beginning and end DOY for each month
        if i_mon[0] > 0:
            beg_day = cum_days_yr[i_mon[0]-1]+1
        else:
            beg_day = 1        
        end_day = cum_days_yr[i_mon[0]]
    
        # Indices for current month
        i_curr_mon = np.where((alldays_yr>=beg_day) & (alldays_yr<=end_day))
        
        # Skip April and later in 2016 (No Data)
        if (i_mon[1]<=3):
            chill_mon[i_yr[0],i_mon[0]] = np.sum(chill[i_curr_mon])
        else:
            if (i_yr[1]<2016):
                chill_mon[i_yr[0],i_mon[0]] = np.sum(chill[i_curr_mon])
                
# Calculate JFM Chill Days
i_jfm  = np.where((mons>=1) & (mons<=3))[0];  chill_JFM  = np.sum(chill_mon[:,i_jfm],axis=1)
i_sond = np.where((mons>=9) & (mons<=12))[0]; chill_SOND = np.sum(chill_mon[:,i_sond],axis=1)

# Plots
sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 1.5})
sns.set_style("darkgrid")
fig = sns.plt.figure(figsize=(12,7))
sns.plt.plot(np.unique(prism_yrs),chill_JFM)
sns.plt.scatter(np.unique(prism_yrs),chill_JFM)
sns.plt.title('Chill Days, JFM ('+site_name+')')
sns.plt.xlim(1981,2016)
sns.plt.ylim(68,92)

sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 1.5})
fig = sns.plt.figure(figsize=(12,7))
sns.plt.plot(np.unique(prism_yrs),chill_SOND)
sns.plt.scatter(np.unique(prism_yrs),chill_SOND)
sns.plt.title('Chill Days, Sep-Dec ('+site_name+')')
sns.plt.xlim(1981,2016)
sns.plt.ylim(40,75)
```
